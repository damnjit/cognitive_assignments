1. For the array gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]'), Find i. Sum of all elements ii. Sum of all elements row-wise iii. Sum of all elements column-wise

```python
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')

total_sum = np.sum(gfg)

row_sum = np.sum(gfg, axis=1)

column_sum = np.sum(gfg, axis=0)

print("Sum of all elements:", total_sum)
print("Sum of elements row-wise:", row_sum)
print("Sum of elements column-wise:", column_sum)
```

2.(a)For the array: array = np.array([10, 52, 62, 16, 16, 54, 453]), find i. Sorted array ii. Indices of sorted array iii. 4 smallest elements iv. 5 largest elements

```python
import numpy as np
my_array = np.array([10, 52, 62, 16, 16, 54, 453])

sorted_array = np.sort(my_array)
sorted_indices = np.argsort(my_array)
smallest_elements = np.sort(my_array)[:4]
largest_elements = np.sort(my_array)[-5:]

print("Sorted array:", sorted_array)
print("Indices of sorted array:", sorted_indices)
print("4 smallest elements:", smallest_elements)
print("5 largest elements:", largest_elements)
```

(b) For the array: array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0]), find i. Integer elements only ii. Float elements only

```python
import numpy as np

array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])

int_elements = array[np.floor(array) == array]
float_elements = array[np.floor(array) != array]

print("Only integer elements:", int_elements)
print("Only float elements:", float_elements)
```

3 You are given a weekly sales dataset and need to perform various operations using NumPy broadcasting. a) Generate your unique sales dataset:  Take the sum of the ASCII values of the initials of your first and last name. Call this value X. (If your initials are A B → ASCII sum = 65

66 = 131 → sales = [131, 181, 231, 281, 331].)  Create a NumPy array sales with values [X, X+50, X+100, X+150, X+200]. b) Compute your personalized tax rate as ((X % 5) + 5) / 100.  Use broadcasting to apply this tax rate to each sales value. c) Adjust sales based on discount:  If sales < X+100, apply a 5% discount.  If sales >= X+100, apply a 10% discount. d) Expand sales data for multiple weeks:  Create a 3×5 matrix representing three weeks of sales by stacking sales three times using broadcasting.  Increase sales by 2% per week using element-wise broadcasting.

```python
import numpy as np

X = ord('R') + ord('G')  # ASCII sum
sales = np.array([X, X+50, X+100, X+150, X+200])

tax_rate = ((X % 5) + 5) / 100
tax_amount = sales * tax_rate

discounted_sales = np.where(sales < X+100, sales * 0.95, sales * 0.90)

weeks = 3
weekly_sales = np.vstack([discounted_sales * (1 + i * 0.02) for i in range(weeks)])

print("Array of sales:", sales)
print("Tax Rate:", tax_rate)
print("Tax Amount:", tax_amount)
print("Discounted Sales:", discounted_sales)
print("Weekly Sales:\n", weekly_sales)
```

4.Generate x values using np.linspace() from -10 to 10 with 100 points. Use each function from the list below and compute y values using NumPy:  Y = x2  Y = sin(x)  Y = ex  Y = log(|x| + 1) Plot the chosen function using Matplotlib. Add title, labels, and grid for clarity.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 100)

y_square = x**2
y_sin = np.sin(x)
y_exp = np.exp(x)
y_log = np.log(np.abs(x) + 1)

plt.figure(figsize=(10, 6))

plt.plot(x, y_square, label=r'y=x^2', color='blue', linewidth=2)
plt.plot(x, y_sin, label=r'y=sin(x)', color='red', linewidth=2)
plt.plot(x, y_exp, label=r'y=e^2', color='green', linewidth=2)
plt.plot(x, y_log, label=r'y=log(|x|+1)', color='purple', linewidth=2)

# Adjust y-axis to better visualize all plots
plt.ylim(-5, 50)

plt.title("Plots of Various Functions", fontsize=14)
plt.xlabel("x values", fontsize=12)
plt.ylabel("y values", fontsize=12)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```
